```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\mercg\Downloads\MP2\spotify.csv")
df
```

```python
df.dropna(inplace=True)
df.isnull().sum() #checking for null values
```

```python
tg = df['playlist_genre'].value_counts()
tg.plot(kind='bar')
plt.title('Top Genres by Song Count')
plt.xlabel('Genre')
plt.ylabel('Song Count')
plt
```

```python
plt.hist(df['duration_ms'], bins=20, edgecolor='black')
plt.xlabel('Song Duration')
plt.ylabel('Occurences')
plt.title('Distribution of Song Durations')
plt
```

```python
audio_features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence','loudness']
correlation_matrix = df[audio_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between the Audio Features')
plt
correlation_matrix
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
kmeans.fit(df[['instrumentalness', 'energy',  'acousticness' , 'liveness', 'speechiness', 'danceability', 'valence','loudness']])
df['cluster'] = kmeans.labels_
sns.scatterplot(data=df, x='playlist_genre', y='loudness', hue='cluster')
plt.title('Genre vs Loudness')
plt.xlabel('Playlist Genre')
plt.ylabel('loudness')
plt
```
